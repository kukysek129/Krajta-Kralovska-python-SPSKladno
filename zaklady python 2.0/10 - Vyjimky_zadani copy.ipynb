# ⚡ Výjimky (Exceptions)

## Co je to výjimka?

**Výjimka** (anglicky *exception*) je chyba, která nastane **během běhu programu**. Když k ní dojde a není ošetřena, program se **okamžitě ukončí** s chybovou hláškou.

```
Traceback (most recent call last):
  File "program.py", line 3, in <module>
    vysledek = 10 / 0
ZeroDivisionError: division by zero
```

### Proč se to děje?

Python čte a provádí kód **řádek po řádku** (je to interpretovaný jazyk). Některé chyby proto odhalí až ve chvíli, kdy na problematický řádek dorazí.

### Dva druhy chyb

| Druh | Kdy nastane | Příklad |
|:---|:---|:---|
| **Syntaktická chyba** (`SyntaxError`) | Python **ani nezačne** provádět řádek – kód je špatně zapsaný | `print("Ahoj"` – chybí `)` |
| **Výjimka za běhu** (`Runtime Error`) | Python **začne provádět** řádek, ale operace selže | `10 / 0` – dělení nulou |

> 💡 **Syntaktické chyby** vám většinou podtrhne editor (červená vlnovka). **Výjimky za běhu** jsou záludnější – projeví se až při spuštění, často jen za určitých podmínek (např. uživatel zadá nesmysl).

## 🛡️ `try` – `except` – zachytávání výjimek

Chyby za běhu můžeme **zachytit** a místo pádu programu na ně **smysluplně reagovat**:

```
┌────────────────────────────────────┐
│  try:                              │
│      kód, který MŮŽE vyvolat chybu │
│                                    │
│  except TypChyby:                  │
│      co udělat, KDYŽ chyba nastane │
└────────────────────────────────────┘
```

```
         try:
           │
           ▼
     ┌──────────┐     BEZ chyby
     │ rizikový │  ──────────────►  pokračuje za except
     │   kód    │
     └──────────┘
           │
        CHYBA!
           │
           ▼
     ┌──────────┐
     │  except  │  ──────────────►  pokračuje za except
     │   blok   │
     └──────────┘
```

> 💡 Pokud zachytíte rodičovskou výjimku (např. `Exception`), zachytíte i **všechny** její potomky. To se hodí, ale i zneužívá – viz dále.

---

## 🔍 Jak číst chybovou hlášku (traceback)

Když program spadne, Python vypíše **traceback** – červený blok textu, který přesně říká **co** se stalo a **kde**. Čte se **odspodu nahoru**:

```
Traceback (most recent call last):       ← 1. "Stalo se něco špatného"
  File "program.py", line 5, in <module> ← 2. KDE – soubor a řádek
    vysledek = 10 / 0                    ← 3. KTERÝ ŘÁDEK – konkrétní kód
ZeroDivisionError: division by zero      ← 4. CO – typ chyby + popis
```

### Postup čtení (odspodu!)

```
  ④  ZeroDivisionError: division by zero    ← CO se stalo?
                                                 → Dělení nulou

  ③      vysledek = 10 / 0                  ← KTERÝ řádek to způsobil?
                                                 → Řádek s dělením

  ②    File "program.py", line 5            ← KDE v kódu?
                                                 → Soubor program.py, řádek 5

  ①  Traceback (most recent call last):     ← Začátek chybového výpisu
```

### Příklad delšího tracebacku (s funkcemi)

```
Traceback (most recent call last):
  File "program.py", line 8, in <module>    ← hlavní program zavolal funkci
    vypocet(10, 0)
  File "program.py", line 3, in vypocet     ← uvnitř funkce nastala chyba
    return a / b
ZeroDivisionError: division by zero         ← samotná chyba
```

> 💡 **Tip:** Při delším tracebacku čtěte vždy **poslední 2–3 řádky** – tam je odpověď na to, co se pokazilo. Řádky nad tím ukazují cestu, kudy se program dostal k chybě.

### Cheat sheet

| Část tracebacku | Co říká | Příklad |
|:---|:---|:---|
| **Poslední řádek** | Typ chyby + popis | `ValueError: invalid literal...` |
| `File "...", line N` | Soubor a číslo řádku | `File "app.py", line 12` |
| Odsazený řádek pod ním | Konkrétní kód, který chybu způsobil | `cislo = int(vstup)` |
| Více `File` bloků | Cesta volání funkcí (poslední = místo chyby) | Hlavní program → funkce → chyba |

````python
# === Základní try-except ===
try:
    cislo = int(input("Zadej číslo: "))
    vysledek = 100 / cislo
    print(f"100 / {cislo} = {vysledek}")
except ZeroDivisionError:
    print("❌ Chyba: Nelze dělit nulou!")
except ValueError:
    print("❌ Chyba: To není platné číslo!")

print("✅ Program pokračuje dál...")
````

> 💡 Můžete mít **více bloků `except`** – každý pro jiný typ chyby. Python provede **první**, který odpovídá.

---

### Uložení výjimky do proměnné (`as`)

Pomocí `as` můžeme výjimku uložit do proměnné a zjistit **podrobnosti o chybě**:

````python
# === except ... as ... ===
try:
    with open("neexistuje.txt", "r") as soubor:
        obsah = soubor.read()
except FileNotFoundError as chyba:
    print(f"❌ Soubor nenalezen!")
    print(f"   Detail: {chyba}")
    print(f"   Typ chyby: {type(chyba).__name__}")
````

---

### ⚠️ Holé `except:` (bez typu) – opatrně!

Pokud napíšete `except:` bez uvedení typu chyby, zachytí se **jakákoliv** výjimka:

````python
# === Holé except – zachytí COKOLIV ===
try:
    a = 5
    b = 0
    c = a / b
except:
    print("Něco se pokazilo...")

print("Program pokračuje")
````

> ⚠️ **Proč je holé `except:` špatný nápad?**
>
> Zachytí i chyby, o kterých **nevíte** – třeba překlep v názvu proměnné (`NameError`). Program pak zdánlivě „funguje", ale **skrytě dělá něco jiného**, než čekáte.
>
> ```python
> try:
>     vysledk = 10 / cislo    # překlep: vysledk místo vysledek
> except:
>     print("Dělení nulou")   # Ale ve skutečnosti je to NameError!
> ```
>
> ✅ **Vždy uvádějte konkrétní typ výjimky**, kterou očekáváte.

---

### Kombinace více výjimek

````python
# === Více except bloků ===
try:
    a = 5
    b = 0
    # b = "a"         # odkomentuj pro TypeError
    c = a / b
except ZeroDivisionError:
    print("❌ Nelze dělit nulou!")
except TypeError:
    print("❌ Špatný typ dat – nelze dělit číslo textem!")
except Exception as e:
    print(f"❌ Neočekávaná chyba: {e}")

print("Program pokračuje")
````

## 🔧 Kompletní `try` blok: `else` a `finally`

Kromě `try` a `except` existují ještě dva volitelné bloky:

```python
try:
    # kód, který MŮŽE vyvolat chybu
except TypChyby:
    # co udělat, KDYŽ chyba nastane
else:
    # provede se POUZE pokud NENASTALA žádná chyba
finally:
    # provede se VŽDY – ať už chyba nastala nebo ne
```

```
         try:
           │
     ┌─────┴─────┐
   CHYBA      BEZ CHYBY
     │            │
     ▼            ▼
  except:       else:
     │            │
     └─────┬──────┘
           ▼
        finally:  ← provede se VŽDY
           │
           ▼
     zbytek programu
```

````python
# === try-except-else-finally ===
try:
    cislo = int(input("Zadej číslo: "))
    vysledek = 100 / cislo
except ZeroDivisionError:
    print("❌ Nelze dělit nulou!")
except ValueError:
    print("❌ To není číslo!")
else:
    # Provede se POUZE pokud NENASTALA chyba:
    print(f"✅ Výsledek: {vysledek}")
finally:
    # Provede se VŽDY:
    print("🏁 Konec výpočtu (vždy se zobrazí)")
````

### Kdy co použít?

| Blok | Provede se kdy | Typické použití |
|:---|:---|:---|
| `try` | Vždy (začátek) | Rizikový kód |
| `except` | Jen při chybě | Reakce na chybu, error message |
| `else` | Jen BEZ chyby | Zpracování úspěšného výsledku |
| `finally` | Vždy (konec) | Úklid – zavření souboru, spojení, ... |

> 💡 `else` a `finally` jsou **volitelné**. V praxi se `finally` hodí hlavně při práci se soubory a databázemi.

## 🚀 `raise` – vlastní vyvolání výjimky

Výjimku můžete **vyvolat sami** pomocí klíčového slova `raise`. To se hodí, když chcete **zakázat** neplatné vstupy do vaší funkce:

````python
# === raise – vyvolání vlastní výjimky ===
def nastav_vek(vek):
    if vek < 0:
        raise ValueError("Věk nemůže být záporný!")
    if vek > 150:
        raise ValueError("Věk nemůže být větší než 150!")
    print(f"Věk nastaven na {vek}")

# Funguje:
nastav_vek(25)

# Vyvolá chybu:
try:
    nastav_vek(-5)
except ValueError as e:
    print(f"❌ Chyba: {e}")
````

## 🔄 Typický vzor – opakované načítání vstupu

Velmi častý vzor: ptáme se uživatele **dokola**, dokud nezadá platnou hodnotu:

````python
# === Opakované načítání s try-except ===
while True:
    try:
        cislo = int(input("Zadej celé číslo: "))
        break  # pokud int() uspěje, vyskočíme z cyklu
    except ValueError:
        print("⚠️ To není celé číslo, zkus to znovu!")

print(f"Zadal jsi: {cislo}")
````

> 💡 Tento vzor kombinuje `while True` + `try-except` + `break`. Je to **profesionální** způsob, jak ošetřit vstup od uživatele.

## 📋 Shrnutí

### Syntaxe `try-except`

```python
try:
    # rizikový kód
except KonkrétníChyba:
    # reakce na chybu
except JináChyba as e:
    # reakce + info z proměnné e
else:
    # jen pokud NENASTALA chyba
finally:
    # VŽDY se provede
```

### Přehled bloků

| Blok | Kdy se provede | Povinný? |
|:---|:---|:---|
| `try` | Vždy (začátek) | ✅ Ano |
| `except` | Při chybě | ✅ Ano (alespoň 1) |
| `else` | Bez chyby | ❌ Volitelný |
| `finally` | Vždy (konec) | ❌ Volitelný |

### Nejčastější výjimky

| Výjimka | Příčina |
|:---|:---|
| `ZeroDivisionError` | Dělení nulou |
| `ValueError` | Špatná hodnota (`int("abc")`) |
| `TypeError` | Špatný typ (`5 + "a"`) |
| `FileNotFoundError` | Soubor neexistuje |
| `IndexError` | Index mimo rozsah |
| `KeyError` | Klíč ve slovníku neexistuje |

### Pravidla

| Pravidlo | |
|:---|:---|
| Vždy uvádějte **konkrétní typ** výjimky | `except ValueError:` ✅ |
| Nepoužívejte holé `except:` | Skrývá chyby ❌ |
| Pro opakovaný vstup: `while True` + `try` + `break` | Standardní vzor ✅ |
| `raise` vyvolá výjimku ve vlastním kódu | Pro validaci vstupů ✅ |

> 🎯 **Zapamatuj si:** Výjimky jsou **normální součást** programování. Dobrý program **nepadá** – zachytí chyby a nabídne uživateli smysluplnou reakci!
